```python
import pandas as pd
import boto3
import numpy as np
import os

pd.set_option('display.max_colwidth', None)
```

```python
s3 = boto3.resource('s3')

img_path = os.path.dirname(os.path.abspath('/Users/yuliejung/dev/ott2/data_preprocessing/img/question/Question_A3.png'))
img_path
```

```python
levels = ['A', 'B', 'C', 'D', 'E']

urls = []

for level in levels:
  for i in range(1, 9):
    filename = 'Question_'
    filename = filename + level + str(i)
  
    try:
      filename = filename + ".png"
      imgPath = img_path + '/' + filename
      
      imgFile = open(imgPath, 'rb')
    except FileNotFoundError as e:
      filename = filename.replace("png", 'jpg')
      imgPath = img_path + '/' + filename

      imgFile = open(imgPath, 'rb')
    
    filetype = 'png' if filename.endswith('png') else 'jpg'
    s3.Bucket('netflixschool').put_object(Key=filename, Body=imgFile, ContentType=f'image/{filetype}')

    url = 'https://netflixschool.s3.amazonaws.com/' + filename
    urls.append(url)
    print(url)
    imgFile.close()
  
    
```
