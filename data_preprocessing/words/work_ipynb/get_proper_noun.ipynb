```python
import requests
import pandas as pd
```

## 고유 명사 여부 판단하는 함수
### 'pos' 항목이 'proper noun'이라면 고유명사

```python
def get_proper_noun(word):
  headers = {'User-Agent':'Chrome/66.0.3359.181'}

  # keyword = str(word)
  url = 'https://wordtype.org/api/senses?term=' + str(word)
  res = requests.get(url, headers=headers)
  result = res.json()
  try:
    word_pos = result[0]['pos']
    # [{
    # '_id': '5748c9b100537df24c97c5cb', 
    # 'lemma': 'Baltimore', 
    # 'pos': 'proper noun', 
    # 'lexicalEntryId': 'WktEN_lexicalEntry_40977', 
    # 'senses': [{'label': '', 'definition': 'A city in central Maryland, USA', 'examples': [], 'relations': []}]
    # }]
    if word_pos.strip() == 'proper noun':
      return True

    return False
  except:
    pass

# TEST CODE
print(get_proper_noun('baltimore'))
```

## 파일 불러오기

```python
path = '../final_datasets/words_df_final_naver.csv'
naver_words = pd.read_csv(path)
naver_words['proper_noun'] = False

display(len(naver_words), naver_words.head())
```

## 크롤링
### 데이터 분석 목적의 서버가 아니라면.. colab에서 돌리는 걸 추천!

```python
path = '../final_datasets/words_df_final_naver_proper_noun'

# 총 32738개이므로 range(0, 33)로 돌려야 함
for i in range(0, 16):
    start = i * 1000
    number = 1000

    df_tmp = naver_words[start:start+number].copy()
    df_tmp['proper_noun'] = df_tmp.Word.map(get_proper_noun)

    
    last = start + number
    df_tmp.to_csv(path+'_'+str(start)+'_'+str(last)+'.csv')
```

```python
path = '../final_datasets/words_df_final_naver_proper_noun'
file_name = path + '_0_1000.csv'
df_result = pd.read_csv(file_name, index_col=False)
df_result.drop(['Unnamed: 0'], axis = 1, inplace = True)

for i in range(1, 33):
    start = i * 1000
    number = 1000
    last = start + number

    file_name = path + '_' + str(start) + '_' + str(last) + '.csv'
    
    df_data = pd.read_csv(file_name, index_col=False)
    df_data.drop(['Unnamed: 0'], axis=1, inplace=True)
    df_result = pd.concat([df_result, df_data])

df_result.reset_index(drop=True, inplace=True)
display(len(df_result), df_result)

df_result.to_csv(path + '.csv', index=False)
```
