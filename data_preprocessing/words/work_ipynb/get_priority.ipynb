```python
import requests
import pandas as pd

def get_priority(word):
  headers = {'User-Agent':'Chrome/66.0.3359.181'}

  keyword = str(word)
  url = 'https://en.dict.naver.com/api3/enko/search?query=' + keyword
  res = requests.get(url, headers=headers)
  result = res.json()
  try:
    print(keyword)
    word_items = result["searchResultMap"]["searchResultListMap"]["WORD"]["items"]
    # if first result priority > 0: that priority
    # else get oxford priority
    # else get priority of first
    priority = -1
    # some result
    if len(word_items) > 0:
      first_item = word_items[0]
      first_item_priority = first_item['priority']
      if first_item_priority > 0:
        priority = first_item_priority
      else:
        ox_items = [item for item in word_items if item["sourceDictnameKO"] == "옥스퍼드 영한사전"]
        if len(ox_items) == 0:
          priority = first_item_priority
        else:
          priority = ox_items[0]['priority']      
    # no result
    return priority
  except:
    pass
```

```python
path = '../missing_priority.csv'
missing_words = pd.read_csv(path, index_col= 'Word')

missing_words
```

```python
missing_words['naver_priority'] = missing_words.index.map(get_priority)
```
