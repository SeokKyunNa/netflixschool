```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 10)
```

```python
words_df = pd.read_csv("../final_datasets/words_df_dday.csv", index_col='Word')
words_df
```

```python
for index, row in words_df.iterrows():
  if pd.isnull(row.category) and not pd.isnull(row.lexile_category):
    words_df.loc[index, 'category'] = row.lexile_category

words_df[words_df['category'] == 'science']
```

```python
naver_0 = lambda x: x.naver_priority == 0
naver_1 = lambda x: x.naver_priority == 1
naver_2 = lambda x: x.naver_priority == 2

lexile_1 = lambda x: x.lexile_grade == 1
lexile_2 = lambda x: x.lexile_grade == 2
lexile_3 = lambda x: x.lexile_grade == 3
lexile_4 = lambda x: x.lexile_grade == 4
lexile_5 = lambda x: x.lexile_grade == 5
lexile_6 = lambda x: x.lexile_grade == 6
lexile_9 = lambda x: x.lexile_grade == 9

lexile_gen = lambda x: x.lexile_category == 'general'
lexile_ss = lambda x: x.lexile_category == 'social_studies'
lexile_math = lambda x: x.lexile_category == 'math'
lexile_sci = lambda x: x.lexile_category == 'science'


ngsl_a1 = lambda x: x.ngsl_sp_freq == 'A1'
ngsl_a2 = lambda x: x.ngsl_sp_freq == 'A2'
ngsl_a3 = lambda x: x.ngsl_sp_freq == 'A3'
ngsl_a4 = lambda x: x.ngsl_sp_freq == 'A4'
ngsl_b1 = lambda x: x.ngsl_sp_freq == 'B1'
ngsl_b2 = lambda x: x.ngsl_sp_freq == 'B2'
ngsl_b3 = lambda x: x.ngsl_sp_freq == 'B3'
ngsl_c1 = lambda x: x.ngsl_sp_freq == 'C1'
ngsl_c2 = lambda x: x.ngsl_sp_freq == 'C2'
ngsl_c3 = lambda x: x.ngsl_sp_freq == 'C3'

awsl = lambda x: x.awsl == 'Y'
toefl = lambda x: x.toefl == 'Y'
tsl = lambda x: x.tsl == 'Y'
bsl = lambda x: x.bsl == 'Y'

cam_a1 = lambda x: x.camb_level == 'A1'
cam_a2 = lambda x: x.camb_level == 'A2'
cam_b1 = lambda x: x.camb_level == 'B1'
cam_b2 = lambda x: x.camb_level == 'B2'
cam_c1 = lambda x: x.camb_level == 'C1'
cam_c2 = lambda x: x.camb_level == 'C2'

ox_a1 = lambda x: x.oxford_level == 'A1'
ox_a2 = lambda x: x.oxford_level == 'A2'
ox_b1 = lambda x: x.oxford_level == 'B1'
ox_b2 = lambda x: x.oxford_level == 'B2'
ox_c1 = lambda x: x.oxford_level == 'C1'
```

```python
def get_word_level(row):
  level = 0
  if naver_0(row) or (lexile_9(row) and lexile_math(row)) or (lexile_9(row) and lexile_sci(row)) or (lexile_6(row) and lexile_sci(row)):
    level = 15
  
  if ngsl_c3(row) or (lexile_6(row) and lexile_math(row)):
    level = 13
  
  if ngsl_c2(row) or ngsl_c1(row) or (lexile_5(row) and lexile_sci(row)):
    level = 12
  
  if ngsl_b2(row) or ngsl_b3(row) or (lexile_9(row) and lexile_ss(row)) or (lexile_6(row) and lexile_ss(row)) or (lexile_5(row) and lexile_math(row)) or (lexile_4(row) and lexile_sci(row)):
    level = 11
  
  if ngsl_a4(row) or ngsl_b1(row) or awsl(row) or toefl(row) or tsl(row) or bsl(row) or (lexile_5(row) and lexile_ss(row)) or (lexile_4(row) and lexile_ss(row)) or cam_c2(row):
    level = 10

  if ngsl_a3(row) or (lexile_9(row) and lexile_gen(row)) or (lexile_4(row) and lexile_math(row)) or (lexile_3(row) and lexile_sci(row)) or (lexile_3(row) and lexile_math(row)) or  ox_c1(row) or cam_c1(row):
    level = 9

  if naver_1(row) or (lexile_6(row) and lexile_gen(row)) or (lexile_3(row) and lexile_ss(row)) or (lexile_2(row) and lexile_ss(row)) or (lexile_2(row) and lexile_sci(row)) or (lexile_1(row) and lexile_sci(row)):
    level = 8

  if (lexile_5(row) and lexile_gen(row)) or cam_b2(row):
    level = 7

  if cam_b1(row):
    level = 6

  if (lexile_4(row) and lexile_gen(row)) or (lexile_3(row) and lexile_gen(row)) or (lexile_2(row) and lexile_gen(row)):
    level = 5
  
  if cam_a2(row):
    level = 4

  if naver_2(row) or (lexile_2(row) and lexile_math(row)) or (lexile_1(row) and lexile_math(row)) or ox_b1(row):
    level = 3
  
  if (lexile_1(row) and lexile_gen(row)):
    level = 2

  if cam_a1(row) or ox_a1(row) or ox_a2(row):
    level = 1

  if level == 0 and row['category'] != 'NN':
    level = 12
  
  return level
    
```

```python
for index, row in words_df.iterrows():
  level = get_word_level(row)
  words_df.loc[index, 'word_level'] = level
```
