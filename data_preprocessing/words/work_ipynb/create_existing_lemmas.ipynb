Create list of lemmas for words that exist in our word list

```python
import pandas as pd

all_lemmas = pd.read_csv('entire_lemmas_complete.csv', index_col='Word')
words_df = pd.read_csv('words_data.csv', index_col='Word')
all_lemmas
```

```python
existing_lemmas = all_lemmas[all_lemmas.index.isin(words_df.index)]
existing_lemmas
```

```python
existing_lemmas.Lemmas = existing_lemmas.Lemmas.str.replace(',', ';')
existing_lemmas
```
