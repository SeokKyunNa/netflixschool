## 파일 불러오기

```python
import os
path = '../script'

# 다큐멘터리(0), 드라마(1), 영화(2) 선택
category = 2
if category == 0:
    path = path + '/Documentary'
elif category == 1:
    path = path + '/Drama'
else:
    path = path + '/Movie'

_list = os.listdir(path)
file_list = []
for file in _list:
    if file.startswith('.'): continue
    file_list.append(file)

print(len(file_list), file_list[:5])
```

## 여기서 작품 선택합니다 꼭!!

```python
# 바로 위 코드에서 file_list를 참고하여 작품 선택
title = file_list[0]

file_name = title + '.WEBRip.Netflix.en[cc]'
print(f'{path}/{title}/{file_name}')
```

```python
file = pd.read_csv(f'{path}/{title}/{file_name}.csv')
df = file
display(df.info(), df)
```

## 자막이 재생되는 시간 계산

```python
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])
df
```

```python
df['duration'] = df['end'] - df['start']
df['duration'] = df['duration'].astype('str')
df['duration'] = df['duration'].str.split(' ').str[2]
df['duration'] = df['duration'].str.split(':').str[2].astype('float')
display(df.info(), df.head())
```

## 필요없는 컬럼 제거

```python
df = df.drop(columns=['start', 'end'])
df.head()
```

## 단어의 개수 계산(띄어쓰기 기준)


```python
df['words_num'] = df['script'].apply(lambda x: len(x.split(' ')))
df
```

# 1초에 몇 개의 단어가 보여지는지 계산

```python
df['wordsPerSec'] = df['words_num'] / df['duration']
df['secPerWord'] = df['duration'] / df['words_num']
df
```
