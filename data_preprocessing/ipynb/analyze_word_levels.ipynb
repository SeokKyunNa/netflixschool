```python
pd.options.display.float_format = '{:.0f}'.format

word_db_df = pd.read_csv("../words/final_datasets/words_df_final_naver_proper_noun.csv")
word_db_df.set_index('Word', inplace=True)
word_db_df
```

```python
lexile_grade = list(word_db_df['naver_priority'].unique())
lexile_grade
```

```python
temp_word_db_df = word_db_df[word_db_df['lexile_grade'] == 1]
temp_word_db_df
```

```python
oxford_level = ['A1','A2', 'B1', 'B2', 'C1']
lexile_grade = [1, 2, 3, 4, 5, 6, 9]
lexile_category = ['social_studies', 'general', 'science', 'math']
ngsl_freq = ['A1','A2', 'A3', 'A4', 'B1','B2','B3', 'C1', 'C2', 'C3', 'C4', 'C5']
ngsl_sp_freq = ['A1','A2', 'A3', 'A4', 'B1','B2','B3', 'C1', 'C2', 'C3']
naver_priority = [2, 1, 0]

```

```python
df = pd.DataFrame()
```

```python
for columns in oxford_level:
    df[f'oxford_level {columns}'] = 1

for columns in lexile_grade:
    df[f'lexile_grade {columns}'] = 1

for columns in lexile_category:
    df[f'lexile_category {columns}'] = 1

df[f'awsl'] = 1
df[f'toefl'] = 1
df[f'tsl'] = 1
df[f'bsl'] = 1

for columns in ngsl_freq:
    df[f'ngsl_freq {columns}'] = 1

for columns in ngsl_sp_freq:
    df[f'ngsl_sp_freq {columns}'] = 1

for columns in naver_priority:
    df[f'naver_priority {columns}'] = 1
```

```python
df.reset_index()
```

```python
columns = []

for i in oxford_level:
    columns.append(['oxford_level', i])

for i in lexile_grade:
    columns.append(['lexile_grade', i])

for i in lexile_category:
    columns.append(['lexile_category', i])

columns.append(['awsl', 'Y'])
columns.append(['toefl', 'Y'])
columns.append(['tsl', 'Y'])
columns.append(['bsl', 'Y'])

for i in ngsl_freq:
    columns.append(['ngsl_freq', i])

for i in ngsl_sp_freq:
    columns.append(['ngsl_sp_freq', i])

for i in naver_priority:
    columns.append(['naver_priority', i])

rows = []

for i in oxford_level:
    rows.append(['oxford_level', i])

for i in lexile_grade:
    rows.append(['lexile_grade', i])

for i in lexile_category:
    rows.append(['lexile_category', i])

rows.append(['awsl', 'Y'])
rows.append(['toefl', 'Y'])
rows.append(['tsl', 'Y'])
rows.append(['bsl', 'Y'])

for i in ngsl_freq:
    rows.append(['ngsl_freq', i])

for i in ngsl_sp_freq:
    rows.append(['ngsl_sp_freq', i])

for i in naver_priority:
    rows.append(['naver_priority', i])
```

```python
for row in rows:
    result = []

    temp_word_db_df = word_db_df[word_db_df[row[0]] == row[1]]

    for column in columns:
        result.append(round(len(temp_word_db_df[temp_word_db_df[column[0]] == column[1]]) / len(temp_word_db_df) * 100,2))

    df.loc[f'{row[0]}, {row[1]}'] = result
```
