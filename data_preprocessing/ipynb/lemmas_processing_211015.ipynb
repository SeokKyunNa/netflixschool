## 단어 파일 불러와서 합치기

```python
# /content/drive/MyDrive/elice_데이터분석_프로젝트/팀_공유_폴더/words/raw_word_data/new/NAWL_1.0_lemmatized_for_research.csv
# /content/drive/MyDrive/elice_데이터분석_프로젝트/팀_공유_폴더/words/raw_word_data/new/NDL_1.0_lemmatized_for_research.csv
file_name1 = f'{camb_word_path}/NAWL_1.0_lemmatized_for_research.csv'
file_name2 = f'{camb_word_path}/NDL_1.0_lemmatized_for_research.csv'

f = open(file_name1)
reader = csv.reader(f)

csv_list = []
for line in reader:
    csv_list.append(line)
f.close()

f = open(file_name2)
reader = csv.reader(f)

for line in reader:
    csv_list.append(line)
f.close()

df = pd.DataFrame(csv_list)
df.rename(columns={0:'Word'}, inplace=True)
df.set_index('Word', inplace=True)
df
```

## 소문자로

```python
for c in df.columns:
  df[c] = df[c].str.lower()

df
```

## df_result에 있는 것과 없는 것으로 나누기

```python
df_right = df_result.join(df, how='right')
df_new = df_right[df_right['Lemmas'].isnull()]
df_dupl = df_right[~df_right['Lemmas'].isnull()]
# display(df_new, df_dupl)
df_new.head()
# df[df.index=='I']
# df_result[df_result.index=='I']
```

## 없던 단어(df_new) 처리

```python
df_new['Lemmas'] = None
df_new.info()
```

```python
processed_df_new = df_new.copy()

def addLemmas(lemmas, col):
  if lemmas==None or str(lemmas)=='':
    if col==None or str(col)=='':
      return ''
    else:
      return col
  else:
    if col==None or str(col)=='':
      return lemmas
    else:
      if str(col) in str(lemmas).split(';'):
        return lemmas
      else:
        return str(lemmas) + ';' + str(col) + ';'

# processed_df_new['Lemmas2'] = processed_df_new.apply(lambda x: x['Lemmas'].str.cat(x[1], sep=';' if x['Lemmas']==None else x[1]), axis=1)
processed_df_new.dropna(how='all', inplace=True)
for i in range(1, 18):
  processed_df_new['Lemmas'] = processed_df_new.apply(lambda x: addLemmas(x['Lemmas'], x[i]), axis=1)
# processed_df_new['Lemmas'] = processed_df_new.apply(lambda x: addLemmas(x['Lemmas'], x[2]), axis=1)
processed_df_new = pd.DataFrame(processed_df_new['Lemmas'])
# processed_df_new['Lemmas'] = processed_df_new.apply(lambda x: x['Lemmas'].split(';'))
len(processed_df_new)
```

## 존재하던 단어(df_dupl) 처리

```python
processed_df_dupl = df_dupl.copy()

def addLemmas(lemmas, col):
  if lemmas==None or str(lemmas)=='':
    if col==None or str(col)=='':
      return ''
    else:
      return col
  else:
    if col==None or str(col)=='':
      return lemmas
    else:
      if str(col) in str(lemmas).split(';'):
        return lemmas
      else:
        return str(lemmas) + ';' + str(col) + ';'

# processed_df_dupl['Lemmas2'] = processed_df_dupl.apply(lambda x: x['Lemmas'].str.cat(x[1], sep=';' if x['Lemmas']==None else x[1]), axis=1)
processed_df_dupl.dropna(how='all', inplace=True)
for i in range(1, 18):
  processed_df_dupl['Lemmas'] = processed_df_dupl.apply(lambda x: addLemmas(x['Lemmas'], x[i]), axis=1)
# processed_df_dupl['Lemmas'] = processed_df_dupl.apply(lambda x: addLemmas(x['Lemmas'], x[2]), axis=1)
processed_df_dupl['Lemmas'] = processed_df_dupl['Lemmas'].str.replace(';;;', ';')
processed_df_dupl['Lemmas'] = processed_df_dupl['Lemmas'].str.replace(';;', ';')

processed_df_dupl = pd.DataFrame(processed_df_dupl['Lemmas'])
display(len(processed_df_dupl), processed_df_dupl)
```

## 합치고 중복 제거해서 마지막 값만 남기기

```python
# 없던 값들 합치기
display(len(df_result), len(processed_df_new))
df_result = pd.concat([df_result, processed_df_new])
display(len(df_result))
```

```python
# 중복됐던 값들 합치기
display(len(df_result), len(processed_df_dupl))
df_result = pd.concat([df_result, processed_df_dupl])
display(len(df_result))
```
