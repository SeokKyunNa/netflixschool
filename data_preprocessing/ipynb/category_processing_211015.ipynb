## 라이브러리

```python
import pandas as pd
import numpy as np
import csv
import os
```

## path 설정

```python
path = '..'
category_word_path = f'{path}/words/raw_word_data/new/hash'
origin_word_path = f'{path}/words/final_datasets'
```

## 결과 데이터프레임 

```python
print(f'{origin_word_path}/words_levels_df_211015.csv')
if os.path.exists(f'{origin_word_path}/words_levels_df_211015.csv'):
    df_result = pd.read_csv(f'{origin_word_path}/words_levels_df_211015.csv', index_col='Word')
    display(df_result)
else:
  print('not exists file');
```

## 단어 파일 불러와서 합치기

```python
files = ['csi', 'legal_terms', 'medical', 'medical2', 'medical3']
# df = pd.read_csv(f'{category_word_path}/{files[0]}.csv', sep='\t')
df_cate = pd.DataFrame()

for i in range(0, 2):
    file_name = f'{category_word_path}/{files[i]}.csv'
    f = open(file_name, 'r', encoding="UTF-8")
    reader = csv.reader(f)
    csv_list = []
    for line in reader:
        csv_list.append(line)
    f.close()
    new_list = []
    for item in csv_list:
        new_list.append(item[0].split('\t')[0])
    df = pd.DataFrame(new_list)
    df.rename(columns={0:'Word'}, inplace=True)
    df['category'] = files[i]
    # df['word_level'] = 7
    # df.set_index('Word', inplace=True)
    # df
    df_cate = pd.concat([df_cate, df])

for i in range(2, 5):
    df = pd.read_csv(f'{category_word_path}/{files[i]}.csv', header=None)
    df.rename(columns={0:'Word'}, inplace=True)
    df['category'] = files[2]
    # df['word_level'] = 7
    df_cate = pd.concat([df_cate, df])

df_cate = df_cate.set_index('Word', drop=True)
display(len(df_cate), df_cate.tail())
```

## 소문자로 변형

```python
df_cate.index = df_cate.index.str.lower()
df_cate
```

## df_result에 있는 것도 없는 것 나누기

```python
df_right = df_result.join(df_cate, how='right')
df_right

df_new = df_right[df_right['word_level'].isnull()]
df_dupl = df_right[~df_right['word_level'].isnull()]

display(df_new, df_dupl)
```
