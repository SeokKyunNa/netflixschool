파일

```python
path = '/content/drive/MyDrive/elice_데이터분석_프로젝트/data/words'
file1 = ['NGSL+1.01+with+SFI.xlsx', 'ngsl_freq.csv']
file2 = ['NGSL+Spoken+1.01.xlsx', 'ngsl_sp_freq.csv']
input = file2[0]
output = file2[1]
filename = input
```

```python
file = pd.read_excel(f'{path}/{filename}')
```

```python
data = file
display(data.info(), data.head())
```

Lemma 컬럼 제외하고 삭제

```python
data = data['Lemma'].to_frame().reset_index(drop=True)
data
# display(data1.iloc[:, 2:],
# data1[list(range(2, 19))])
# data1[2] = data1.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
# data2[2] = data2.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
# data3[2] = data3.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
# data1
```

Frequency 컬럼에 범주형(Category) 데이터 입력

```python
bins = [0, 999, 1999, 2999, 3999, 5999, 7999, 9999, 14999, 19999, 24999, 29999, 34999]
group_names = ['A1', 'A2', 'A3', 'A4', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3', 'C4', 'C5']
data['Frequency'] = pd.cut(data.index, bins, labels=group_names, include_lowest = True)
```

```python
display(data.head(), data[995:1005], data[9995:10005], data.tail())
```

컬럼명 수정

```python
data = data.rename(columns={'Lemma': 'Word'})
```

파일로 내보내기

```python
# path = '/content/drive/MyDrive/elice_데이터분석_프로젝트/data/words'
data.to_csv(f'{path}/{output}', index=False)
# data.to_csv(f'{path}/ngsl_sp_freq.csv', index=False)
```
