자막이 재생되는 시간 계산

```python
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])
df
```

```python
df['duration'] = df['end'] - df['start']
df
```

```python
df['duration'] = df.apply(lambda x: df['duration'].astype('str'))
df['duration'] = df['duration'].str.split(' ').str[2]
df
```

```python
df['duration'] = df['duration'].str.split(':').str[2].astype('float')
df
```

필요없는 컬럼 제거

```python
df = df.drop(columns=['start', 'end'])
df
```

단어의 개수 계산(띄어쓰기 기준)


```python
df['script_num'] = df['script'].apply(lambda x: len(x.split(' ')))
df
```

1초에 몇 개의 단어가 보여지는지 계산

```python
df['scriptPerSec'] = df['duration'] / df['script_num']
df
```
