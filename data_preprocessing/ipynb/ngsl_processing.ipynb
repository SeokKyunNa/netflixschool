라이브러리

```python
import pandas as pd
```

파일

```python
path = '/content/drive/MyDrive/elice_데이터분석_프로젝트/data/words'
filename = 'NGSL+1.01+by+band.xlsx'
```

```python
data1 = pd.read_excel(f'{path}/{filename}', sheet_name=0, header=None)
data2 = pd.read_excel(f'{path}/{filename}', sheet_name=1, header=None)
data3 = pd.read_excel(f'{path}/{filename}', sheet_name=2, header=None)
display(data1.info(), data1.head())
```

NaN 데이터들을 공백문자('')로 채워줌

```python
data1 = data1.fillna('')
data2 = data2.fillna('')
data3 = data3.fillna('')
data1.head()
```

컬럼 2에 그 뒤의 값들을 모으고 세미콜론을 구분자로 넣어줌

```python
# display(data1.iloc[:, 2:],
# data1[list(range(2, 19))])
data1[2] = data1.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
data2[2] = data2.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
data3[2] = data3.iloc[:, 2:].agg(';'.join, axis=1).str.rstrip(';')
data1
```

컬럼 3부터는 필요없으니 삭제

```python
data1 = data1.iloc[:, 1:3]
data2 = data2.iloc[:, 1:3]
data3 = data3.iloc[:, 1:3]
data1
```

데이터프레임 합치기 (data1, data2, data3)

```python
result = pd.concat([data1, data2, data3]).reset_index(drop=True)
display(len(result), result)
```

컬럼명 변경

```python
result = result.rename(columns={1:'Word', 2:'Lemmas'})
result
```

파일로 내보내기

```python
# path = '/content/drive/MyDrive/elice_데이터분석_프로젝트/data/words'
result.to_csv(f'{path}/ngsl_with_lemmas.csv', index=False)
```
