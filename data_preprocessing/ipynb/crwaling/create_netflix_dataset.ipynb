```python
import pandas as pd
import numpy as np
```

```python
netflix = pd.read_csv('../../../server/csv_files/netflix_titles.csv')
netflix

netflix.show_id = np.nan
netflix
```

```python
netflix2 = pd.read_csv('../../../server/csv_files/netflix_titles 2.csv')
netflix2
```

```python
netflix_contents = pd.concat([netflix2, netflix])
netflix_contents

netflix_contents = netflix_contents.drop_duplicates(subset=['title', 'type', 'release_year', 'director'], keep='first')
netflix_contents['show_id'] = netflix_contents['show_id'].fillna(0).astype(int)
netflix_contents.sort_values('title', inplace=True)
netflix_contents

netflix_contents['country'].unique().tolist()

# drop countries of non-english speaking films
english = ['United States', 'United Kingdom', 'Australia', 'New Zealand', 'Bahamas', 'Canada', 'Dominica', 'Ireland', 'Jamaica', 'Malta']

def checkEnglish(countries):
  countries_split = str(countries).split(',')
  for item in countries_split:
    if item in english:
      return True
  return False
english_mask = netflix_contents['country'].apply(checkEnglish)
netflix_contents = netflix_contents[english_mask]

netflix_contents

# Drop columns
netflix_contents = netflix_contents.drop(columns=['cast', 'date_added', 'description', 'country'])
netflix_contents = netflix_contents.rename(columns={'title': 'title_en', 'rating': 'age_rating', 'duration': 'running_time', 'listed_in': 'genre', 'show_id': 'netflix_id'})
netflix_contents.set_index('netflix_id', inplace=True)
netflix_contents
```
