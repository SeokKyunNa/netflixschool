## 라이브러리

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
```

## path 설정

```python
path = '../..'
# path = 
word_path = f'{path}/words/final_datasets'
```

## 파일 불러오기

```python
df = pd.read_csv(f'{word_path}/all_scripts.csv').set_index('title')
df
```

## level_1 ~ level_6까지 weight 적용

```python
df_calculated = df.copy()
for i in range(1, 7):
    level = 'level_' + str(i)
    df_calculated[level] = df_calculated[level] * df_calculated['weight']

df_calculated
```

```python
df_calculated.describe()
```

## 정규화

```python
df_calculated_level = df_calculated[['level_1', 'level_2', 'level_3', 'level_4', 'level_5', 'level_6']]
df_normalization = (df_calculated_level - df_calculated_level.min()) / (df_calculated_level.max() - df_calculated_level.min())
df_normalization
```

```python
df_result = df_normalization.join(df[['WPS_mean','WPS_std', 'WPS_min', 'WPS_max']])
df_result
```

## csv 파일로 변환

```python
df_result.to_csv(f'{word_path}/all_scripts_normalization.csv')
```
