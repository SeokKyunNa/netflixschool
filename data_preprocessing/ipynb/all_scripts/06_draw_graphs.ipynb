## 라이브러리

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## path 설정

```python
path = '../..'
csv_path = f'{path}/script/csv_files'
```

## 파일 불러오기

```python
file_name = 'all_scripts_final_score.csv'
df = pd.read_csv(f'{csv_path}/{file_name}', index_col='title')
df
```

## 4사분면 그래프 그리기

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x='level_score', y='wps_score', color='r')
plt.title('Netflix Contents English Difficulty Distribution', size=20, pad=20)

for i in range(df.shape[0]):
    plt.text(df.level_score[i], y=df.wps_score[i], s=df.index[i], alpha=0.7)

plt.axhline(y=2, color='k', linewidth=1, linestyle='-')
plt.text(x=10.2, y=4.1, s='WPS', weight='bold')
plt.axvline(x=10, color='k', linewidth=1, linestyle='-')
plt.text(x=20, y=1.9, s='word level', weight='bold')

plt.axis('off')

plt.show()

# Scatter Plot (Image by author)
```
