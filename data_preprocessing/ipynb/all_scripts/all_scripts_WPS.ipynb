# 위의 코드들 합쳐서 한 번에 모든 작품 돌리기

```python
import os
import pandas as pd
import numpy as np
# import seaborn as sns
# import matplotlib.pyplot as plt
# import matplotlib as mpl
# import warnings
# warnings.filterwarnings('ignore')

path = '../..'
# path = '/content/drive/MyDrive/elice_데이터분석_프로젝트/팀_공유_폴더'
script_path = f'{path}/script'
word_path = f'{path}/words/final_datasets'

if os.path.exists(f'{word_path}/all_scripts_WPS.csv'):
    df_result = pd.read_csv(f'{word_path}/all_scripts_WPS.csv', index_col='title')
else:
    df_result = pd.DataFrame(columns=['title', 'mean', 'std', 'min', 'max']).set_index('title')

display(df_result)

# 다큐멘터리(0), 드라마(1), 영화(2) 선택
category = 2
if category == 0:
    script_path_cat = script_path + '/Documentary'
elif category == 1:
    script_path_cat = script_path + '/Drama'
else:
    script_path_cat = script_path + '/Movie'

_list = os.listdir(script_path_cat)
file_list = []
for file in _list:
    if file.startswith('.'): continue
    file_list.append(file)

file_list = sorted(file_list)
print(f'작품개수: {len(file_list)}')
count = 0
for index, title in enumerate(file_list):
    print(f'[{index}: {title}]', end=', ')
    count += 1
    if count%5 == 0:
        print()
```

```python
for i in range(len(file_list)):
    # 바로 위 코드에서 file_list를 참고하여 작품 선택
    title = file_list[i]
    if (df_result.index==title).any(): continue

    file_name = title + '.WEBRip.Netflix.en[cc]'
    # print(f'{path}/{title}/{file_name}')

    df_file = pd.read_csv(f'{script_path_cat}/{title}/{file_name}.csv')
    df = df_file

    df['start'] = pd.to_datetime(df['start'])
    df['end'] = pd.to_datetime(df['end'])

    df['duration'] = df['end'] - df['start']
    df['duration'] = df['duration'].astype('str')
    df['duration'] = df['duration'].str.split(' ').str[2]
    df['duration'] = df['duration'].str.split(':').str[2].astype('float')

    df = df.drop(columns=['start', 'end'])

    df['words_num'] = df['script'].apply(lambda x: len(x.split(' ')))

    df['wordsPerSec'] = df['words_num'] / df['duration']
    df['secPerWord'] = df['duration'] / df['words_num']

    df = df[df['words_num']!=1]
    df = df[df['wordsPerSec']>=0.5].sort_values('words_num')

    result = df['wordsPerSec'].describe().to_frame().transpose()[['mean', 'std', 'min', 'max']]
    result['title'] = title
    result = result.set_index('title')

    df_result = pd.concat([df_result, result])

df_result
```

## CSV 파일로 변환

```python
df_result.to_csv(f'{word_path}/all_scripts_WPS.csv')
```
