## 라이브러리

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
```

## path 설정

```python
path = '../..'
# path = 
script_path = f'{path}/script/csv_files'
```

## 파일 불러오기

```python
df = pd.read_csv(f'{script_path}/all_scripts_normalization.csv').set_index('title')
df
```

## level 별 가중치 계산

```python
df_result = df.copy()
level_1 = df_result['level_1'] * 1
level_2 = df_result['level_2'] * 2
level_3 = df_result['level_3'] * 3
level_4 = df_result['level_4'] * 4
level_5 = df_result['level_5'] * 5
level_6 = df_result['level_6'] * 6

df_result['level_score'] = level_1 + level_2 + level_3 + level_4 + level_5 + level_6
df_result
```

## WPS Score 계산

```python
df_result['wps_score'] = df_result['WPS_mean_nlln'] * 4
df_result
```

## Final Score 계산

```python
df_result['final_score'] = df_result['level_score'] + df_result['wps_score']
df_result
```

## csv 파일로 변환

```python
df_result.to_csv(f'{script_path}/all_scripts_final_score.csv')
```
